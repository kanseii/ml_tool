```python
import pandas as pd
```

```python
fandango = pd.read_csv('fandango_score_comparison.csv')
print(type(fandango))
fandango_films = fandango.set_index('FILM',drop=False)
print()
fandango_films
```

```python
fandango_films["Avengers: Age of Ultron (2015)":"Hot Tub Time Machine 2 (2015)"] # slice 切片
fandango_films.loc["Avengers: Age of Ultron (2015)":"Hot Tub Time Machine 2 (2015)"] # 结果跟上面一样
fandango_films.loc['Kumiko, The Treasure Hunter (2015)'] # 具体的电影
movies = ['Kumiko, The Treasure Hunter (2015)', 'Do You Believe? (2015)', 'Ant-Man (2015)'] # 一个list
fandango_films.loc[movies]
```

```python
import numpy as np
types = fandango_films.dtypes
# print(types)
float_columns = types[types.values == 'float64'].index
float_df = fandango_films[float_columns]

fun = lambda x: np.std(x)    # 参考https://www.cnblogs.com/xiaxuexiaoab/p/9226772.html
deviations = float_df.apply(fun) # 计算标准差

print(deviations)
```

```python
rt_mt_user = float_df[['RT_user_norm', 'Metacritic_user_nom']]
rt_mt_user.apply(lambda x: np.std(x), axis=1)   # 0 for row, 1 for col
```
